10. Perform a simulation in which people enter a premises according to an exponential distribution with parameter lambda = 1 person/minute, and determine how many people enter the premises in an hour

 In each iteration, one person must enter. People have to enter (continue iterating) until the accumulated time exceeds one hour (60 minutes). Then you must quantify the total number of people who entered the premises.\
 You should then iterate this many times and calculate the average number of people


 Do you find any relationship with the results of previous exercises?

```python
acum = []

for i in range(10000):
    contador = 0
    hora = 0
    while hora < 60:
        hora_entrada = np.random.exponential(1)
        hora += hora_entrada
        contador += 1
    acum.append(contador)
```

```python
sns.countplot(x = acum)
plt.xticks(rotation = "vertical")
plt.show()
```

11. Perform a simulation to estimate the daily turnover of a store. The following data is available:
 *The work day has 8 hours
 * Clients arrive according to an exponential distribution of 30 clients per hour
 * The spending of each customer in the store follows a normal distribution of average 50 euros and deviation 10 euros

```python
# 8 horas
# lam = 30
# x~N(50,10)
acum=[]
for i in range(10000):
    clientes_por_dia = np.random.poisson(lam = 30, size=8)
    gasto_clientes = np.random.normal(loc = 50, scale = 10,size = sum(clientes_por_dia))
    facturado = sum(gasto_clientes)
    acum.append(facturado)
```

12. Graph the probability density function of the turnover calculated in the previous item

```python
sns.histplot(acum)
```

## Calculate probabilities

### Highs

1. Graph a distribution that represents the heights of a population, the average height is 1.7 meters and the standard deviation is 0.1 meters.

```python
numeros = np.random.normal(loc = 170, scale = 10, size = 10000)
sns.histplot(numeros)
```

2. Calculate the probability that a person is taller than 2 meters


```python
personas_altas = len(numeros[numeros > 200])
personas_altas / len(numeros) * 100
```

2. Calculate the probability that a person is less than 1.4 meters tall

```python
personas_altas = len(numeros[numeros < 140])
personas_altas / len(numeros) * 100
```

3. Calculate the probability that a person measures between 1.5 meters and 1.7 meters

```python
personas_altas = len(numeros[(numeros > 150)&(numeros < 170)])
personas_altas / len(numeros) * 100
```

4. Calculate the probability that a person measures exactly 1.7 meters

```python
personas_altas = len(numeros[numeros == 170])
personas_altas / len(numeros) * 100
```

5. Calculate the probability that a person is not between 1.4 meters and 2 meters tall.

```python
personas_altas = len(numeros[~((numeros > 140)&(numeros < 200))])
personas_altas / len(numeros) * 100
```

### Roulette

1. Graph a probability distribution showing the number of times I will get red on the roulette wheel (win) by spinning it 20 times. Remember that there are 18 red numbers and 37 numbers

```python
n = 20
p = 18/37

numeros=np.random.binomial(n, p,100000)
sns.countplot(x=numeros, color = "blue")
plt.show()
```

2. Calculate the probability of having won 20 times

```python
personas_altas = len(numeros[numeros == 20])
personas_altas / len(numeros) * 100
```

3. Calculate the probability of winning more than half of the time

```python
personas_altas = len(numeros[numeros > 10])
personas_altas / len(numeros) * 100
```

4. Calculate the probability of having lost all the rounds

```python
personas_altas = len(numeros[numeros == 0])
personas_altas / len(numeros) * 100
```

5. Calculate the probability of having won an even number of times (2, 4, 6, ..., 20)

```python
personas_altas = len(numeros[numeros % 2 == 0])
personas_altas / len(numeros) * 100
```
