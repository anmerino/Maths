```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numeros = np.random.normal(loc=20,scale = 3,size = 10000000)
```

```python
numeros_2 = np.random.normal(loc=20,scale = 3/8,size = 10000000)
```

```python
plt.hist(numeros,bins = 2000,alpha = 0.8)
plt.vlines(x = 20,ymin=0,ymax=22000,colors = "red")
plt.title("Tiempo de entrega Pizzeria")
plt.xlabel("minutos")
plt.yticks([])

plt.xlim((10,30))
```

```python
plt.hist(numeros,bins = 2000,alpha = 0.2)

plt.hist(numeros_2,bins = 1500, color = "forestgreen")
plt.vlines(x = 20,ymin=0,ymax=30000,colors = "red")

plt.xlim((10,30))
plt.title("Tiempo de entrega Pizzeria")
plt.xlabel("minutos")
plt.yticks([])

plt.show()
```

```python
plt.hist(numeros_2,bins = 1500, color = "forestgreen")
plt.vlines(x = 20,ymin=0,ymax=30000,colors = "red")

plt.title("Tiempo de entrega Pizzeria")
plt.xlabel("minutos")
plt.yticks([])

plt.show()
```

```python
plt.hist(numeros_2[numeros_2<20.617],bins = 950, color = "forestgreen")
plt.hist(numeros_2[numeros_2>=20.617],bins = 470,color = "red")
plt.vlines(x = 20,ymin=0,ymax=30000,colors = "red")
plt.title("Tiempo de entrega Pizzeria")
plt.xlabel("minutos")
plt.yticks([])
plt.show()
```
